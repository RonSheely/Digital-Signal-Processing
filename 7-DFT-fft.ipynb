```python
#generate an appridiocsignal.....
import numpy as np
import matplotlib.pyplot as plt
F = 200
Fs=3000
L= 512
x = np.zeros(L)
for n in range(L):
 x[n] = np.sin(2*np.pi*350*n/Fs) + np.sin(2*np.pi*550*n/Fs) + np.sin(2*np.pi*900*n/Fs)
#print(x)
plt.plot(x[1:L//2])
#plt.stem(x)
```

```python
N = L;
X= np.zeros(N, dtype = 'complex_')
for k in range(N):
    for n in range(N):
        X[k] = X[k] + x[n]*np.exp(-1j*2*np.pi*k*n/N)
plt.plot((abs(X)**2))
```

```python
#plt.plot((abs(c)**2))
plt.stem((abs(X)**2))
```

```python
Y = np.fft.fft(x,N)
plt.plot((abs(X)**2))
```
