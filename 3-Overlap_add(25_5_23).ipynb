```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#signal generation
F1 = 100
F2 = 200
F3 = 350
F4 = 600
Fs = 2000
SL = 2000

s = np.zeros(SL)
for n in range(SL):
    s[n] = np.sin(2*np.pi*F1*n/Fs)+np.sin(2*np.pi*F2*n/Fs)+np.sin(2*np.pi*F3*n/Fs)+np.sin(2*np.pi*F4*n/Fs)
plt.plot(s[0:512])
```

```python
# DFT

def DFT(x,N):
    X = np.zeros(N,dtype = 'complex_')
    for k in range(N):
        for n in range(N):
            X[k] = X[k] + x[n]*np.exp(-1j*2*np.pi*k*n/N)
    return X
```

```python
# IDFT
def IDFT(X, N):
    y = np.zeros(N,dtype = 'complex_')
    for n in range(N):
        for k in range(N):
            y[n] = y[n] + X[k]*np.exp(1j*2*np.pi*k*n/N)
    
    return y/N
```
