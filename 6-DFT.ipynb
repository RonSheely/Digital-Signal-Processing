```python
#generate a sinusoid.....
import numpy as np
import matplotlib.pyplot as plt
F = 10
Fs=600
L= 300
x = np.zeros(L)
for n in range(L):
 x[n] = np.sin(2*np.pi*F*n/Fs) + np.sin(2*np.pi*2*F*n/Fs) + np.sin(2*np.pi*3*F*n/Fs)+ np.sin(2*np.pi*4*F*n/Fs)
#print(x)
plt.plot(x)
#plt.stem(x)

```

```python
N = 300;
c = np.zeros(L, dtype = 'complex_')
for k in range(L):
    for n in range(L):
        c[k] = c[k] + x[n]*np.exp(-1j*2*np.pi*k*n/N)

plt.plot(c)
```
