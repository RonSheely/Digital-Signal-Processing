```python
#generate an aperiodic signal
import numpy as np
import matplotlib.pyplot as plt
F = 2
Fs = 40
L = 100
x = np.zeros(L)
for n in range(L):
  x[n] = np.sin(2*np.pi*F*n/Fs)
plt.plot(x[0:L//2])
```

```python
wn = np.random.normal(0,1,L) 

wn = wn/np.max((wn))
y = x + wn

#plt.plot(wn[0:L])
plt.plot(y[0:L//2])
```

```python
M = 7
s = np.zeros(L)
for i in range(L-M):
  s[i] = sum(y[i:i+M])/M
plt.plot(s[0:L//2])  
```
